```python
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
```

```python
# Can see all the API Get Requests available here: https://github.com/man-c/pycoingecko

#Simple get price, volume, market cap, 24hr change
cg.get_price(ids='bitcoin,litecoin,ethereum', vs_currencies='usd,eur', include_market_cap=True, include_24hr_vol=True, include_24hr_change=True, include_last_updated_at=True)
```

```python
# Timestamps - not required?
import time
import datetime
start_date = "01/01/2019"
end_date = "31/12/2019"

UNIX_start = time.mktime(datetime.datetime.strptime(start_date, "%d/%m/%Y").timetuple())
UNIX_end = time.mktime(datetime.datetime.strptime(end_date, "%d/%m/%Y").timetuple())

print(f"Start Date = {start_date}, or in UNIX {UNIX_start}")
print(f"End Date = {end_date}, or in UNIX {UNIX_end}")
```

```python
# /coins/{id}/history (Get historical data (name, price, market, stats) at a given date for a coin)
cg.get_coin_history_by_id(id='bitcoin', date='30-12-2017')
```

```python
# /coins/{id}/market_chart/range (Get historical market data include price, market cap, and 24h volume within a range of timestamp (granularity auto))
# need to convert UNIX back to normal dates and bring into a DataFrame
cg.get_coin_market_chart_range_by_id(id='bitcoin', vs_currency='usd', from_timestamp = UNIX_start, to_timestamp = UNIX_end)
```

```python
# Alternatively look up by day, but then cannot pick specific dates
cg.get_coin_market_chart_by_id(id='bitcoin', vs_currency='usd', days=500)
```
